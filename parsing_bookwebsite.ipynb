```python
#–ü–∞—Ä—Å–µ—Ä, –∫–æ—Ç–æ—Ä—ã–π –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π, –≥–¥–µ –Ω–∞ –ø–µ—Ä–≤–æ–º –º–µ—Å—Ç–µ ‚Äì¬†–æ—Ç–∑—ã–≤, –∞ –Ω–∞ –≤—Ç–æ—Ä–æ–º ‚Äì –æ—Ü–µ–Ω–∫–∞

def mybookparser(url, pages):

  data = []

  for i in range(1, pages+1):
      page = requests.get(url +f'?page={i}')
      soup = BeautifulSoup(page.text, 'html')

      revs = []
      review = soup.find_all('article')
      for r in reviews:
          revs.append(r.text)

      stars = []
      for ul_tag in soup.find_all('ul', class_='ant-rate'):
        for li_tag in ul_tag.find_all('li'):
                stars.append(li_tag['class'][1].split('-')[-1])

      stars_digits = [1 if x == 'full' else 0 for x in stars] #words into numbers
      #–∫–æ–¥ —Å–æ–∑–¥–∞–µ—Ç —Å–ø–∏—Å–æ–∫, –≥–¥–µ –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç ‚Äî —ç—Ç–æ —Å—É–º–º–∞ –ø—è—Ç–∏ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∏–∑ stars_digits
      stars_for_reviews = [sum(stars_digits[idx:idx + 5]) for idx in range(0, len(stars_digits), 5)]
      for idx in range(0, len(revs)):
          data.append((revs[idx], stars_for_reviews[idx]))
  sleep(3)
  return data
```

```python
def mybookparser(url, pages):

  data = [] #–∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö. –ø—É—Å—Ç—ã–µ —Å–ø–∏—Å–∫–∏ –¥–ª—è —Å–æ—Ö—Ä –¥–∞–Ω–Ω—ã—Ö
  headers ={

  }

  for i in range(1, pages +1): #–ø—Ä–æ—Ö–æ–¥ –ø–æ 1 —Å—Ç—Ä –¥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ pages
      try: #–æ–±—Ä–∞–±–æ—Ç–∫–∞ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü—ã
          page = requests.get(url +f'?page={i}', headers=headers)
          page.raise_for_status()
          soup = BeautifulSoup(page.text, 'html.parser') #–ø–∞—Ä—Å–∏–Ω–≥ html
          revs = [] #—Å–±–æ—Ä –æ—Ç–∑—ã–≤–æ–≤. —Ç—É–¥–∞ –¥–æ–±–∞–≤–ª—è—é—Ç—Å—è –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã article
          reviews = soup.find_all('article')
          for r in reviews:
              revs.append(r.text.strip())

          stars = [] #—Å–±–æ—Ä –æ—Ü–µ–Ω–æ–∫
          for ul_tag in soup.find_all('ul', class_='ant-rate'):
              for li_tag in ul_tag.find_all('li'):
                  stars.append(li_tag['class'][1].split('-')[-1])

          stars_digits = [1 if x == 'full' else 0 for x in stars] #–ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –æ—Ü–µ–Ω–æ–∫. –ø—É—Å—Ç–∞—è –∑–≤–µ–∑–¥–∞ - 0, –∑–≤–µ–∑–¥–∞ - 1
          stars_for_reviews = [sum(stars_digits[idx:idx + 5]) for idx in range(0, len(stars_digits), 5)]

          for idx in range(0, len(revs)): #—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ –∏—Ç–æ–≥–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö.
          #–∫–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ –æ—Ç–∑—ã–≤–∞ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–æ–±–∞–≤–ª—è—é—Ç—Å—è –≤ —Å–ø–∏—Å–æ–∫ data –∫–∞–∫ –∫–æ—Ä—Ç–µ–∂
              data.append((revs[idx], stars_for_reviews[idx]))


          print(f"–£—Å–ø–µ—à–Ω–æ {i}/{pages}")
      except Exception as e:
          print(f"–û—à–∏–±–∫–∏ {i}: {e}")
      sleep(3)  # Avoid overwhelming the server

  return data
```

```python
#–ó–∞–≥—Ä—É–∑–∫–∞ RuBERT –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ —Ç–æ–Ω–∞–ª—å–Ω–æ—Å—Ç–∏
model_name = "sismetanin/rubert-ru-sentiment-rusentiment"
try:
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name)
    print("RuBERT model loaded successfully")
except Exception as e:
    print(f"Error loading RuBERT model: {e}")
    exit()
```

```python
def predict_sentiment(text):
    try:
        inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512) #–ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –æ—Ç–∑—ã–≤ + –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω—ã–π —Ç–æ–∫–µ–Ω–∏–∑–∞—Ç–æ—Ä, –∫–æ—Ç–æ—Ä—ã–π –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç —Ç–µ–∫—Å—Ç
        with torch.no_grad(): #–±–µ–∑ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –≥—Ä–∞–¥–∏–µ–Ω—Ç–æ–≤ (–æ–ø—Ç–∏–º–∏–∑–∏—Ä–æ–≤–∞–Ω–æ –¥–ª—è –ø–∞–º—è—Ç–∏)
            outputs = model(**inputs) #–≤—ã–∑—ã–≤–∞–µ—Ç –º–æ–¥–µ–ª—å, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –Ω–∞ –æ—Å–Ω–æ–≤–µ –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –≤–≤–æ–¥–∞
        logits = outputs.logits #–Ω–µ–æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã–µ –≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –æ—Ç –º–æ–¥–µ–ª–∏, –∫–æ—Ç –ø—Ä–µ–æ–±—Ä–∞–∑—É—é—Ç—Å—è –≤ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ `softmax`(–Ω–æ—Ä–º–∞–ª–∏–∑—É–µ—Ç –ª–æ–≥–∏—Ç—ã, —á—Ç–æ–±—ã —Å—É–º–º–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –±—ã–ª–∞ —Ä–∞–≤–Ω–∞ 1)
        probabilities = torch.softmax(logits, dim=1)
        label_id = torch.argmax(probabilities, dim=1).item() #–Ω–∞—Ö–æ–¥–∏—Ç –∏–Ω–¥–µ–∫—Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –≤ –º–∞—Å—Å–∏–≤–µ probabilities, –∫–æ—Ç–æ—Ä—ã–π —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–æ–º—É –∫–ª–∞—Å—Å—É (—Ç–æ–Ω–∞–ª—å–Ω–æ—Å—Ç–∏)
        labels = ["negative", "neutral", "positive", "speech", "skip"]
        return labels[label_id], probabilities[0].tolist()
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞: {text[:50]}... Error: {e}")
        return "error", [0.0] * 5


tqdm.pandas()
df[['sentiment', 'probabilities']] = df['review'].progress_apply(predict_sentiment).apply(pd.Series)
print("–ê–Ω–∞–ª–∏–∑ —Ç–æ–Ω–∞–ª—å–Ω–æ—Å—Ç–∏:")
print(df[['review', 'star', 'sentiment', 'rev_len']].head())
```
