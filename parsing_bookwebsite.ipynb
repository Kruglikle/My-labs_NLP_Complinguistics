```python
#Парсер, который возвращает список кортежей, где на первом месте – отзыв, а на втором – оценка

def mybookparser(url, pages):

  data = []

  for i in range(1, pages+1):
      page = requests.get(url +f'?page={i}')
      soup = BeautifulSoup(page.text, 'html')

      revs = []
      review = soup.find_all('article')
      for r in reviews:
          revs.append(r.text)

      stars = []
      for ul_tag in soup.find_all('ul', class_='ant-rate'):
        for li_tag in ul_tag.find_all('li'):
                stars.append(li_tag['class'][1].split('-')[-1])

      stars_digits = [1 if x == 'full' else 0 for x in stars] #words into numbers
      #код создает список, где каждый элемент — это сумма пяти последовательных значений из stars_digits
      stars_for_reviews = [sum(stars_digits[idx:idx + 5]) for idx in range(0, len(stars_digits), 5)]
      for idx in range(0, len(revs)):
          data.append((revs[idx], stars_for_reviews[idx]))
  sleep(3)
  return data
```

```python
model_name = "sismetanin/rubert-ru-sentiment-rusentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
def mybookparser(url, pages):

  data = [] #инициализация данных. пустые списки для сохр данных
  headers ={

  }

  for i in range(1, pages +1): #проход по 1 стр до количества pages
      try: #обработка каждой страницы
          page = requests.get(url +f'?page={i}', headers=headers)
          page.raise_for_status()
          soup = BeautifulSoup(page.text, 'html.parser') #парсинг html
          revs = [] #сбор отзывов. туда добавляются все элементы article
          reviews = soup.find_all('article')
          for r in reviews:
              revs.append(r.text.strip())

          stars = [] #сбор оценок
          for ul_tag in soup.find_all('ul', class_='ant-rate'):
              for li_tag in ul_tag.find_all('li'):
                  stars.append(li_tag['class'][1].split('-')[-1])

          stars_digits = [1 if x == 'full' else 0 for x in stars] #преобразование оценок. пустая звезда - 0, звезда - 1
          stars_for_reviews = [sum(stars_digits[idx:idx + 5]) for idx in range(0, len(stars_digits), 5)]

          for idx in range(0, len(revs)): #формирование итоговых данных.
          #каждая строка отзыва и соответствующая оценка добавляются в список data как кортеж
              data.append((revs[idx], stars_for_reviews[idx]))


          print(f"Успешно {i}/{pages}")
      except Exception as e:
          print(f"Ошибки {i}: {e}")
      sleep(3)  # Avoid overwhelming the server

  return data
```

```python
print("Ready")
url = "https://mybook.ru/author/aleksandr-dyuma/graf-monte-kristo/reviews/"
pages = 60
our_url = mybookparser(url, pages)
print(f"Collected {len(our_url)} reviews")
```

```python
df = pd.DataFrame(our_url, columns=['review', 'star'])
df['rev_len'] = df['review'].apply(len)
print("Создано:")
print(df.head())
```

```python
#Загрузка RuBERT для анализа тональности
model_name = "sismetanin/rubert-ru-sentiment-rusentiment"
try:
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name)
    print("RuBERT model loaded successfully")
except Exception as e:
    print(f"Error loading RuBERT model: {e}")
    exit()
```

```python
def predict_sentiment(text):
    try:
        inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512) #принимает на вход отзыв + предобученный токенизатор, который обрабатывает текст
        with torch.no_grad(): #без вычисления градиентов (оптимизировано для памяти)
            outputs = model(**inputs) #вызывает модель, которая предсказывает результаты на основе подготовленных данных ввода
        logits = outputs.logits #необработанные выходные данные от модели, кот преобразуются в вероятности с помощью функции `softmax`(нормализует логиты, чтобы сумма вероятностей была равна 1)
        probabilities = torch.softmax(logits, dim=1)
        label_id = torch.argmax(probabilities, dim=1).item() #находит индекс максимальной вероятности в массиве probabilities, который соответствует предсказанному классу (тональности)
        labels = ["negative", "neutral", "positive", "speech", "skip"]
        return labels[label_id], probabilities[0].tolist()
    except Exception as e:
        print(f"Ошибка: {text[:50]}... Error: {e}")
        return "error", [0.0] * 5


tqdm.pandas()
df[['sentiment', 'probabilities']] = df['review'].progress_apply(predict_sentiment).apply(pd.Series)
print("Анализ тональности:")
print(df[['review', 'star', 'sentiment', 'rev_len']].head())
```

```python
output_path = '/content/mybookreviews_analyzed.csv'
df.to_csv(output_path, index=False, encoding='utf-8')
print(f"Results saved to {output_path}")
```
